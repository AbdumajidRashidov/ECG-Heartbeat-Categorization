```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
train_df = pd.read_csv('./data/mitbih_train.csv', header=None)
test_df = pd.read_csv('./data/mitbih_test.csv', header=None)

# Basic Data Exploration
print(train_df.describe())
print(train_df.info())

# Plotting histograms for the first few features
train_df.iloc[:, :10].hist(figsize=(15, 10), bins=20)
plt.suptitle("Feature Distributions")
plt.show()

# Class Distribution
class_counts = train_df.iloc[:, -1].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

# Correlation Analysis
correlation_matrix = train_df.iloc[:, :-1].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title("Correlation Matrix of Features")
plt.show()

# Time Series Analysis: Plotting ECG signals for a few samples from each class
classes = train_df.iloc[:, -1].unique()
plt.figure(figsize=(15, 10))
for i, cls in enumerate(classes):
    plt.subplot(len(classes), 1, i + 1)
    plt.plot(train_df[train_df.iloc[:, -1] == cls].iloc[0, :-1])
    plt.title(f"Class {cls} ECG Signal")
plt.tight_layout()
plt.show()
```
